```python
# Haversine function
def haversine(lat1, lon1, lat2, lon2):
    R = 6371  # Radius of Earth in km
    lat1, lon1, lat2, lon2 = map(np.radians, [lat1, lon1, lat2, lon2])
    
    dlat = lat2 - lat1
    dlon = lon2 - lon1
    
    a = np.sin(dlat / 2.0)**2 + np.cos(lat1) * np.cos(lat2) * np.sin(dlon / 2.0)**2
    c = 2 * np.arctan2(np.sqrt(a), np.sqrt(1 - a))
    
    return R * c  # Distance in km

# Apply the function row-wise
fdf['distance'] = fdf.apply(lambda row: haversine(row['start_lat'], 
                                                row['start_lng'], 
                                                row['end_lat'], 
                                                row['end_lng']), 
                                                axis=1)


```

```python
# Define seasons
def find_season(month):
        season_month = {
            12:'Winter', 1:'Winter', 2:'Winter',
            3:'Spring', 4:'Spring', 5:'Spring',
            6:'Summer', 7:'Summer', 8:'Summer',
            9:'Autumn', 10:'Autumn', 11:'Autumn'}
        return season_month.get(month)
```

```python
# Create columns for seasons
season_list = []
for month in fdf['month']:
    season = find_season(month)
    season_list.append(season)
    
fdf['season'] = season_list

# Convert the new Season column to categorical type
fdf["season"] = pd.Categorical(fdf["season"])
```
