```python
import geopandas as gpd
import pandas as pd
from shapely.geometry import Point# Load the wards GeoJSON file
wards_gdf = gpd.read_file("wards_from_2022.geojson")
```

```python
wards_gdf 
```

```python
# Ensure the GeoDataFrame has a proper coordinate reference system (CRS) for area calculation
wards_gdf = wards_gdf.to_crs(epsg=3395)  # Use a projected CRS suitable for area calculation

# Calculate the area in square meters and convert to square kilometers
wards_gdf['area_km2'] = wards_gdf['geometry'].area / 1e6

# Reset the CRS back to the original for further operations
wards_gdf = wards_gdf.to_crs(epsg=4326)

print(wards_gdf[['WARD', 'area_km2']])
```

```python
# Merge unique_stations with ward areas
unique_stations_with_area = unique_stations.merge(wards_gdf[['WARD', 'area_km2']], left_on='start_ward', right_on='WARD')

# Calculate density
unique_stations_with_area['density'] = unique_stations_with_area['unique_stations'] / unique_stations_with_area['area_km2']

# Display the result
print(unique_stations_with_area[['start_ward', 'year', 'density']])
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set figure size & style
plt.figure(figsize=(15, 8))
sns.set_style("whitegrid")

# Create barplot with better colors
sns.barplot(
    data=unique_stations_with_area, 
    x='start_ward', 
    y='density', 
    hue='year', 
    palette="Blues",  # Try 'magma', 'crest', or 'Set2' for variety
    edgecolor='black',  # Clearer bar edges
    dodge=True  # Ensures hue separation
)


# Title & labels
plt.title('Density of Unique Stations by Ward and Year (2021-2023)', fontsize=16, fontweight='bold')
plt.xlabel('Ward', fontsize=14)
plt.ylabel('Density (Unique Stations per kmÂ²)', fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

# Improve legend
plt.legend(title='Year', title_fontsize=13, fontsize=12, loc="upper right")

# Show the plot
plt.show()
```

```python
# Combine the dataframes for all years
combined_df = pd.concat([fdf_gdf21, fdf_gdf22, fdf_gdf23])

# Melt the dataframe to consider both start_ward and end_ward
melted_df = pd.melt(combined_df, id_vars=['ride_id', 'year'], value_vars=['start_ward', 'end_ward'], 
                    var_name='ward_type', value_name='ward')

# Group by ward and year, and count the number of rides
rides_by_ward_year = melted_df.groupby(['ward', 'year']).size().reset_index(name='ride_count')

# Calculate the total rides per year
total_rides_per_year = rides_by_ward_year.groupby('year')['ride_count'].transform('sum')

# Calculate the share of rides
rides_by_ward_year['ride_share'] = (rides_by_ward_year['ride_count'] / total_rides_per_year) * 100

# Display the result
print(rides_by_ward_year)
```

```python
# Set figure size and style
plt.figure(figsize=(15, 8))
sns.set_style("whitegrid")

# Create barplot for ride share
sns.barplot(
    data=rides_by_ward_year, 
    x='ward', 
    y='ride_share', 
    hue='year', 
    palette="Blues",  # Color palette for better visualization
    edgecolor='black',   # Softer look
    width=0.8,        # Clearer bar edges# Clearer bar edges
    dodge=True  # Ensures hue separation         # Ensures hue separation
)

# Title and labels
plt.title('Share of Rides per Ward and Year (2021-2023)', fontsize=16, fontweight='bold')
plt.xlabel('Ward', fontsize=14)
plt.ylabel('Ride Share (%)', fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

# Improve legend
plt.legend(title='Year', title_fontsize=13, fontsize=12, loc="upper right")

# Show the plot
plt.show()
```
