```python
import matplotlib.pyplot as plt
import seaborn as sns

# List of columns related to commuting
commuting_columns = [
    'commuting_carpooled', 'commuting_drove_alone', 'commuting_other_means', 
    'commuting_public_transportation', 'walked_to_work', 'worked_from_home'
]

# Convert 'ward' to numeric for sorting (assuming 'Ward X' format)
wards_eco_2['ward_num'] = wards_eco_2['ward'].str.extract(r'(\d+)').astype(int)
wards_eco_2 = wards_eco_2.sort_values('ward_num')  # Sort by numeric ward value

# Set up the plot (adjust number of rows/cols dynamically if needed)
num_wards = len(wards_eco_2)
rows = (num_wards // 4) + (num_wards % 4 > 0)  # Create enough rows
fig, axes = plt.subplots(nrows=rows, ncols=4, figsize=(18, rows * 5))  # More spacing
axes = axes.flatten()  # Flatten for easy iteration

# Get Set2 colors
colors = sns.color_palette("Set2", n_colors=len(commuting_columns))

# Loop through each ward and create a pie chart
for i, row in enumerate(wards_eco_2[['ward', *commuting_columns]].iterrows()):
    ward = row[1]['ward']
    commuting_data = row[1][commuting_columns].astype(float).values  # Convert values to float

    # Define labels for the pie chart
    labels = ['Carpooled', 'Drove Alone', 'Other Means', 'Public Transport', 'Walked', 'Worked from Home']

    # Plot the pie chart with Set2 colors
    wedges, texts, autotexts = axes[i].pie(
        commuting_data, labels=labels, autopct='%1.1f%%', startangle=90, 
        colors=colors, textprops={'fontsize': 10}  # Apply Set2 colors
    )  

    axes[i].set_title(f'Commuting Habits in {ward}', fontsize=12)
    axes[i].axis('equal')  # Keep the pie chart circular

# Hide unused subplots (if any)
for j in range(i + 1, len(axes)):
    fig.delaxes(axes[j])

# Increase spacing for better readability
plt.subplots_adjust(hspace=0.3, wspace=0.9)
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.ticker as ticker

# Create figure
plt.figure(figsize=(12, 7))

# Bar plot with improved color palette
sns.barplot(x="ward", 
            y="mean_household_income", 
            data=wards_eco_2, 
            hue='ward',
            palette="Blues_r", 
            width=0.8)  # Keeping bars wide for clarity

# Add value labels on bars
for p in plt.gca().patches:
    plt.text(p.get_x() + p.get_width() / 2, 
            p.get_height(), 
            f"${p.get_height():,.0f}".replace(',', '.'),  # Formatting with dots as thousand separator
            ha='center', va='bottom', fontsize=10, fontweight='bold')
    
    
# Function to format y-axis labels with USD and dots as thousand separators
def format_with_dollar(x, _):
    return f'${x:,.0f}'.replace(',', '.')

# Apply the formatter to the y-axis
plt.gca().yaxis.set_major_formatter(ticker.FuncFormatter(format_with_dollar))

# Titles and labels
plt.title('Mean Household Income by Ward (2019-2023)', fontsize=16)
plt.xlabel('Ward', fontsize=12)
plt.ylabel('Per Capita Income (USD)', fontsize=12)

# Layout adjustment
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.ticker as ticker

# Employment Rate by Ward
plt.figure(figsize=(12, 7))

# Create barplot
sns.barplot(x="ward", 
            y="employment_rate", 
            data=wards_eco_2,
            hue="ward",
            palette="Blues_r", 
            width=0.8)  # Keeping bars wide for clarity

# Add % symbol to bar labels
for p in plt.gca().patches:
    plt.text(p.get_x() + p.get_width() / 2, 
             p.get_height(), 
             f"{p.get_height():.1f}%",  # Just adding '%' symbol
             ha='center', va='bottom', fontsize=10, fontweight='bold')

# Function to format y-axis labels with dots as thousand separators
def format_with_dots(x, _):
    return f'{x:,.0f}%'.replace(',', '.')  # Keeping format consistent

# Apply the formatter to the y-axis
plt.gca().yaxis.set_major_formatter(ticker.FuncFormatter(format_with_dots))

plt.title('Employment Rate by Ward (2019-2023)', fontsize=16)
plt.xlabel('Ward', fontsize=12)
plt.ylabel('Employment Rate (%)', fontsize=12)  # Explicitly mention percentage

plt.tight_layout()
plt.show()
```
