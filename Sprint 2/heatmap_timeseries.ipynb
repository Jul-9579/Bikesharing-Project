```python
import numpy as np
import folium
from folium import plugins

initial_data = np.random.normal(size=(100, 2)) * np.array([[1, 1]]) + np.array(
    [[48, 5]]
)

move_data = np.random.normal(size=(100, 2)) * 0.01

data = [(initial_data + move_data * i).tolist() for i in range(100)]
```

```python
time_ = 0
N = len(data)
itensify_factor = 30
for time_entry in data:
    time_ = time_+1
    for row in time_entry:
        weight = min(np.random.uniform()*(time_/(N))*itensify_factor, 1)
        row.append(weight)
```

```python
import json

print(json.dumps(data, indent=2))
```

```python
f = folium.Figure(width=800, height=500)
m = folium.Map([48.0, 5.0], zoom_start=6).add_to(f)

hm = plugins.HeatMapWithTime(data).add_to(m)

m
```

```python
m.save("heatmap.html")
```

```python
from datetime import datetime, timedelta

time_index = [(datetime.now() + k * timedelta(1)).strftime("%Y-%m-%d") for k in range(len(data))]

time_index
```

```python
f = folium.Figure(width=1000, height=500)
m = folium.Map([48.0, 5.0], zoom_start=6).add_to(f)

plugins.HeatMapWithTime(data, index=time_index, auto_play=True, max_opacity=0.3).add_to(m)

m
# m.save("index.html")
```
