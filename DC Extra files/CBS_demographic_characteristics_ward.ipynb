```python
# Create new broader age groups
wards_demo['age_under_18'] = wards_demo[['age_under_5', 'age_5_to_9', 'age_10_to_14', 'age_15_to_19']].sum(axis=1)
wards_demo['age_18_to_44'] = wards_demo[['age_20_to_24', 'age_25_to_34', 'age_35_to_44']].sum(axis=1)
wards_demo['age_45_to_64'] = wards_demo[['age_45_to_54', 'age_55_to_59', 'age_60_to_64']].sum(axis=1)
wards_demo['age_65_and_over'] = wards_demo[['age_65_to_74', 'age_75_to_84']].sum(axis=1)

# Display the first few rows to verify
wards_demo[['ward', 'age_under_18', 'age_18_to_44', 'age_45_to_64', 'age_65_and_over']].head(10)
```

### Age Distribution
Wards with higher share of rides (Ward 2, 1, 6) have the highest share of population in the age 18 to 44 group. The average age of CaBi users being 34 (https://drive.google.com/file/d/1KRbFgQGX50Zo4lzgNPySsQVoLk8wGee8/view), this age factor can explain the higher bike usage. Ward 7 has one of the lowest proportion of population in the age 18 to 44 group.

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Define the new age groups
new_age_columns = ['age_under_18', 'age_18_to_44', 'age_45_to_64', 'age_65_and_over']

# Sort wards_demo by ward number
wards_demo_sorted = wards_demo.sort_values(by='ward')

# Define a visually appealing color palette
custom_colors = sns.color_palette("Set2", len(new_age_columns))  # Use 'viridis' for vibrant colors

# Create subplots for pie charts
fig, axes = plt.subplots(2, 4, figsize=(20, 10))  # Adjust grid size based on the number of wards
axes = axes.flatten()

# Loop through each ward and create a pie chart
for i, (index, row) in enumerate(wards_demo_sorted[['ward'] + new_age_columns].iterrows()):
    # Extract ward and age distribution values
    ward = row['ward']
    age_values = row[new_age_columns].values  # Extract age group values

    # Plot pie chart with custom colors
    axes[i].pie(
        age_values, 
        labels=new_age_columns, 
        autopct='%1.1f%%', 
        startangle=90, 
        colors=custom_colors  # Apply custom colors
    )
    axes[i].set_title(f'Age Distribution - {ward}', fontsize=14, fontweight='bold')  

# Adjust layout for better spacing
plt.tight_layout()

# Show the plot
plt.show()
```

```python
# Merge the 'native', 'pacific_islander', and 'mixed_race' columns into a new column 'other_race'
wards_demo['other_race'] = wards_demo[['native', 'pacific_islander', 'mixed_race']].sum(axis=1)

# Drop the original columns if they are no longer needed
wards_demo.drop(columns=['native', 'pacific_islander', 'mixed_race'], inplace=True)

# Display the first few rows to verify
wards_demo.head()
```

```python
# Ensure the column renaming is applied to wards_demo_sorted
wards_demo_sorted.rename(columns=column_renames, inplace=True)

# Ensure the 'other_race' column is present in wards_demo_sorted
wards_demo_sorted['other_race'] = wards_demo['other_race']

# Define the columns for race distribution
race_columns = [
    'latino',
    'white',
    'black',
    'asian',
    'other_race'
]

# Define a visually appealing color palette
custom_colors = sns.color_palette("Set2", len(race_columns))  # Use 'Set3' for diverse colors

# Create subplots for pie charts
fig, axes = plt.subplots(2, 4, figsize=(20, 10))  # Adjust grid size based on the number of wards
axes = axes.flatten()

# Loop through each ward and create a pie chart
for i, (index, row) in enumerate(wards_demo_sorted[['ward'] + race_columns].iterrows()):
    # Extract ward and race distribution values
    ward = row['ward']
    race_values = row[race_columns].values  # Extract race group values

    # Plot pie chart with custom colors
    axes[i].pie(
        race_values, 
        labels=race_columns, 
        autopct='%1.1f%%', 
        startangle=90, 
        colors=custom_colors  # Apply custom colors
    )
    axes[i].set_title(f'Race Distribution - {ward}', fontsize=14, fontweight='bold')

# Adjust layout for better spacing
plt.tight_layout()

# Show the plot
plt.show()
```
